2. Run Perceptron Learning Algorithm (PLA) on the data set above. Report the number
of updates that PLA takes before converging. Plot the examples {(xn, yn)}, the target
function f , and the final hypothesis g in the same figure. Comment on whether f is
close to g

```
arr = [0, 0, 0]
def fit(data):
    updates = 0
    while True: 
        mistakes = 0
        updates += 1
        for arg_1,arg_2, arg_3 in data:
            y_pred = predict(arg_1,arg_2)
            if y_pred == 0 :
                y_pred = -1
            if arg_3 == 0:
                arg_3 = -1
            if arg_3 != y_pred:
                mistakes += 1
                arr[1] = arr[1] + arg_3 * arg_1
                arr[2] = arr[2] + arg_3 * arg_2
                arr[0] = arr[0] + arg_3
        if mistakes == 0:
            break
    print("Done! Number of updates: " + str(updates))

def predict(arg_1, arg_2):
    if arr[0] + arr[1] * arg_1 + arr[2] * arg_2 < 0:
        return 0
    return 1

fit(data)
```

```
plt.plot([arg_1 for (arg_1,arg_2, arg_3) in data if arg_3==0],
        [arg_2 for (arg_1,arg_2, arg_3) in data if arg_3==0], '<', mfc='none', mec='r')

plt.plot([arg_1 for (arg_1,arg_2, arg_3) in data if arg_3==1],
         [arg_2 for (arg_1,arg_2, arg_3) in data if arg_3==1], '>', mfc='none',  mec='g')

plot(X_data[:,0],-val1[0]*X_data[:,0]/val1[1]-val2/val1[1]);
def f(x):
    return -(arr[0] + arr[1]*x)/arr[2]
dim = np.arange(-10, 10)
plt.plot(dim, [f(x) for x in dim]);
/*











*/
```
